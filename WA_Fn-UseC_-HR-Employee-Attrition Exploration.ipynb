```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

# View the age distribution of the employees

```python
plt.figure(figsize=(20,5))
sns.barplot(df['Age'].value_counts())
plt.title('Age distribution of the employees')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

# explore data for Attrition by Age

```python
plt.figure(figsize=(20,5))
sns.countplot(x=df['Age'],hue=df['Attrition'])
plt.title('explore data for Attrition by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

# explore data for Left employees breakdown

```python
plt.figure(figsize=(20,5))
plt.pie(df['Attrition'].value_counts(), labels = ('No','Yes'),autopct = '%1.2f%%',explode = (0.1,0))
plt.title('explore data for Left employees breakdown')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

# Explore data for Education Field distribution

```python
plt.figure(figsize=(20,5))
plt.pie(df['EducationField'].value_counts(),autopct='%1.1f%%',labels=tuple(dict(df['EducationField'].value_counts()).keys()))
plt.title('Explore data for Education Field distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

# explore data for Marital Status

```python
plt.figure(figsize=(20,5))
sns.barplot(df['MaritalStatus'].value_counts())
plt.title('explore data for Marital Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
